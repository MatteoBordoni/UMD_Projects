### From Wide to Long Sample

```python
long_sample = pd.melt(wide_sample, id_vars=['placekey', 'city', 'region', 'date_range_start', 'date_range_end'], 
             value_vars=['visits_by_day_1', 'visits_by_day_2', 'visits_by_day_3', 'visits_by_day_4', 
                         'visits_by_day_5', 'visits_by_day_6', 'visits_by_day_7'], 
             var_name='day_of_week', value_name='daily_visits')
long_sample.head()
```

```python
long_sample.head()
```

```python
long_sample['day_of_week'] = long_sample['day_of_week'].str[-1:].astype(int)
long_sample['date'] = pd.to_datetime(long_sample['date_range_start']) + pd.to_timedelta(long_sample['day_of_week'] - 1, unit='d')
long_sample = long_sample[['placekey', 'city', 'region', 'date', 'daily_visits']]
long_sample.head()
```

```python
long_sample['daily_visits'].describe()
```

Let's see whether the variable contains outliers or not:

```python
data_mean, data_std = mean(long_sample['daily_visits']), std(long_sample['daily_visits'])

cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off
outliers = [x for x in long_sample['daily_visits'] if x < lower or x > upper]
print('Identified outliers: %d' % len(outliers))

outliers_removed = [x for x in long_sample['daily_visits'] if x >= lower and x <= upper]
print('Non-outlier observations: %d' % len(outliers_removed))
```

## Data Summary 

```python
yearly_long_sample = long_sample.groupby(long_sample['date'].dt.year).agg({'date': 'min',
                                           'daily_visits': 'count' })
yearly_long_sample['average'] = long_sample.groupby(long_sample['date'].dt.year)['daily_visits'].mean()
yearly_long_sample = yearly_long_sample.rename(columns = {'date' : 'begin_date', 'daily_visits' : '#_of_visits'})
yearly_long_sample['end_date'] = long_sample.groupby(long_sample['date'].dt.year)['date'].max()
yearly_long_sample = yearly_long_sample.reset_index()
yearly_long_sample = yearly_long_sample[['date', 'begin_date', 'end_date', '#_of_visits', 'average']]
yearly_long_sample.head()
```

```python
long_sample['day_of_week'] = long_sample['date'].dt.day_name()
long_sample.head()
```

```python
long_sample['positive_visits'] = np.where(long_sample['daily_visits'] > 0, 1, 0)
long_sample.head()
```

```python
long_sample['day_of_week'] = pd.Categorical(long_sample['day_of_week'], categories=['Monday', 
                                                                                    'Tuesday', 
                                                                                    'Wednesday',
                                                                                    'Thursday',
                                                                                    'Friday', 
                                                                                    'Saturday', 
                                                                                    'Sunday'], ordered=True)
tabular = long_sample.groupby(['day_of_week', 
        'positive_visits']).size().unstack(fill_value=0)[[0, 1]].reset_index().sort_values(by='day_of_week')
tabular
```
